### <a name='7'></a> Ćwiczenie 7
Do poniższego obiektu _quotations_:

```
stocks = {'PLW': 387.00, 'CDR': 339.5, 'TEN': 349.5, '11B': 391.0}
quotations = pd.Series(data=stocks)
```
Dodaj dwa elementy:
* indeks: BBT, wartość: 25.5
* indeks: F51, wartość: 19.2

__Oczekiwany wynik:__


```
PLW    387.0
CDR    339.5
TEN    349.5
11B    391.0
BBT     25.5
F51     19.2
dtype: float64
```

__Wskazówka:__ Użyj funkcji _pd.concat()._



````python
stocks = {'PLW': 387.00, 'CDR': 339.5, 'TEN': 349.5, '11B': 391.0}
quotations = pd.Series(data=stocks)
````

````python
quotations
````

````python
pd.concat([quotations, pd.Series({'BBT': 25.5, 'F51': 19.2})])
````

### <a name='8'></a> Ćwiczenie 8

### <a name='9'></a> Ćwiczenie 9

|  0 | Amazon    |  2375   | AMZN.US  |
